```python
from pydot.pydot import *
```

```python
# http://www.graphviz.org/doc/info/attrs.html
shape = 'note' # http://www.graphviz.org/doc/info/shapes.html

g = Dot()
br = Node('Baranyi Roberts 1994', shape=shape, style="filled", fillcolor='white')
br_mr = Node('Baranyi 1997 (m=r)', shape=shape, style="filled", fillcolor='white')
br_nu = Node('Baranyi Roberts 1994 (&#957;=1)', shape=shape, style="filled", fillcolor='white')
br_nu_mr =  Node('Baranyi 1997 (&#957;=1, m=r)', shape=shape, style="filled", fillcolor='white')
br_m0 =  Node('Richards 1959 (1/m=0)', shape=shape, style="filled", fillcolor='white')
br_nu_m0 =  Node('Logistic (&#957;=1, 1/m=0)', shape=shape, style="filled", fillcolor='white')

p6 = Subgraph('cluster_6', label='6 parameters', style="filled", fillcolor='#aaaaaa', rank=6)
p5 = Subgraph('cluster_5', label='5 parameters', style="filled", fillcolor='#aaaaaa', rank=5)
p4 = Subgraph('cluster_4', label='4 parameters', style="filled", fillcolor='#aaaaaa', rank=4)
p3 = Subgraph('cluster_3', label='3 parameters', style="filled", fillcolor='#aaaaaa', rank=3)
g.add_subgraph(p6)
g.add_subgraph(p5)
g.add_subgraph(p4)
g.add_subgraph(p3)

p6.add_node(br)
p5.add_node(br_mr)
p5.add_node(br_nu)
p4.add_node(br_nu_mr)
p4.add_node(br_m0)
p3.add_node(br_nu_m0)

# nu = 1
g.add_edge(Edge(br, br_nu, label=' &#957;=1'))
g.add_edge(Edge(br_mr, br_nu_mr, label=' &#957;=1'))
g.add_edge(Edge(br_m0, br_nu_m0, label=' &#957;=1'))
# m = r
g.add_edge(Edge(br, br_mr, label=' m=r'))
g.add_edge(Edge(br_nu, br_nu_mr, label=' m=r'))
# 1/m = 0
g.add_edge(Edge(br, br_m0, label=' 1/m=0'))
g.add_edge(Edge(br_nu, br_nu_m0, label=' 1/m=0'))

g.set_dpi(300)
g.set_layout('dot')
g.set_rankdir('LR')
# g.set_splines('compound')
g
```

```python
g.write_pdf('model_map.pdf')
!open model_map.pdf
```
